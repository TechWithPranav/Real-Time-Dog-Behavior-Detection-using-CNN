```python

from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Dropout, Flatten, Dense
import os
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
TRAIN_DIR = 'images/train'
TEST_DIR = 'images/test'
```

```python
# Define the data generators with augmentation
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
)
```

```python
test_datagen = ImageDataGenerator(rescale=1./255)
```

```python
train_generator = train_datagen.flow_from_directory(
    TRAIN_DIR,
    target_size=(384, 384),
    batch_size=64,
    class_mode='categorical'
)
```

```python
test_generator = test_datagen.flow_from_directory(
    TEST_DIR,
    target_size=(384, 384),
    batch_size=64,
    class_mode='categorical'
)
```

```python
# Define the model with reduced complexity
model = Sequential()
model.add(Conv2D(32, kernel_size=(3, 3), activation='relu', input_shape=(384, 384, 3)))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.25))

model.add(Conv2D(64, kernel_size=(3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.25))

model.add(Conv2D(128, kernel_size=(3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.25))

model.add(Flatten())
model.summary()
```

```python
# Define the model with reduced complexity
model = Sequential()
model.add(Conv2D(32, kernel_size=(3, 3), activation='relu', input_shape=(384, 384, 3)))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.25))

model.add(Conv2D(64, kernel_size=(3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.25))

model.add(Conv2D(128, kernel_size=(3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.25))

model.add(Flatten())

model.add(Dense(256, activation='relu'))
model.add(Dropout(0.5))

model.add(Dense(4, activation='softmax'))
```

```python
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
# Determine the number of batches in the training and validation sets
train_steps_per_epoch = train_generator.samples // train_generator.batch_size
validation_steps_per_epoch = test_generator.samples // test_generator.batch_size

# Train the model using data generators
model.fit(
    train_generator,
    steps_per_epoch=train_steps_per_epoch,
    epochs=100,
    validation_data=test_generator,
    validation_steps=validation_steps_per_epoch
)
```

```python
model_json = model.to_json()
with open("dog_emotion3.json",'w') as json_file:
    json_file.write(model_json)
model.save("dog_emotion3.h5")
```

```python
from keras.models import model_from_json
import numpy as np
```

```python
json_file = open("dog_emotion3.json", "r")
model_json = json_file.read()
json_file.close()
model = model_from_json(model_json)
model.load_weights("dog_emotion3.h5")
```

```python
label = ['angry','happy','sad','relaxed']
```

```python
from keras_preprocessing.image import load_img,img_to_array
```

```python
def ef(image_path, color_mode='rgb'):
    # Load the image with specified color mode
    img = load_img(image_path, color_mode=color_mode)
    
    # Convert the image to a NumPy array
    feature = np.array(img)
    
    # Check if the image is in grayscale or RGB
    if color_mode == 'grayscale':
        target_shape = (384, 384, 1)  # Grayscale image has one channel
    else:
        target_shape = (384, 384, 3)  # RGB image has three channels
    
    # Reshape the array to the desired input shape
    feature = feature.reshape(1, *target_shape)
    
    # Normalize pixel values
    return feature / 255.0
```

```python
image = 'images/train/sad/116854_10d7d8031d_b.jpg'
print("original image is of sad")
img = ef(image)
pred = model.predict(img)
pred_label = label[pred.argmax()]
print("model prediction is ",pred_label)
```
